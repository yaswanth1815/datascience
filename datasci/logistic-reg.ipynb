```python
drop=['fractal_dimension_mean','texture_se','symmetry_se']
x1=x.drop(drop,axis=1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
def forward_selection(a, b, model, max_features=None):
    selected_features = []
    remaining_features = list(a.columns)
    best_score = 0.0
    if max_features is None:
        max_features = len(remaining_features)
    for i in range(max_features):
        scores_with_candidates = []
        for feature in remaining_features:
            current_features = selected_features + [feature]
            X_train, X_test, y_train, y_test = train_test_split(a[current_features], y, test_size=0.2, random_state=42)
            model.fit(X_train, y_train)
            y_pred = model.predict(X_test)
            score = accuracy_score(y_test, y_pred)
            scores_with_candidates.append((score, feature))
        scores_with_candidates.sort(reverse=True)
        best_new_score, best_new_feature = scores_with_candidates[0]
        if best_new_score > best_score:
            best_score = best_new_score
            selected_features.append(best_new_feature)
            remaining_features.remove(best_new_feature)
            print(f"Selected: {best_new_feature}, Accuracy: {best_new_score:.4f}")
        else:
            break
    
    return selected_features
model = LogisticRegression(solver='liblinear', multi_class='ovr')
selected_features = forward_selection(x1, y, model)
print("Final selected features:", selected_features)
```
