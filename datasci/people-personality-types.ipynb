```python
import numpy as np
import pandas as pd
import tensorflow as tf
import sklearn
import matplotlib.pyplot as plt
import seaborn as sb
from sklearn.preprocessing import LabelEncoder
```

```python
data=pd.read_csv('/kaggle/input/predict-people-personality-types/data.csv')
data=pd.DataFrame(data)
data.head()
```

```python
le=LabelEncoder()
data['Gender']=data['Gender'].map({'Male':1,'Female':0,})
data['Interest']=data['Interest'].map({'Unknown':0, 'Sports':1, 'Others':2, 'Technology':3, 'Arts':4})
data['personality']=le.fit_transform(data['Personality'])
```

```python
data2=data.drop('Personality',axis=1)
data2
```

```python
data2.isnull().sum()
```

```python
data2.describe
```

```python
data2.shape
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
from sklearn.model_selection import train_test_split
x=data2.drop('Thinking Score',axis=1)
y=data2['Thinking Score']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=49)
```

```python
x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_test,y_pred)
```

```python
model.score(x_train,y_train)
```

```python
from sklearn.linear_model import Ridge
model2=Ridge(alpha=2)
model2.fit(x_train,y_train)
y_pred2=model2.predict(x_test)
```

```python
model2.score(x_train,y_train)
mean_absolute_error(y_pred2,y_test)
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
poly = PolynomialFeatures(degree=2) 
X_poly = poly.fit_transform(x_train)

model3 = LinearRegression()
model3.fit(X_poly, y_train)
y_pred3 = model3.predict(poly.transform(x_test))
```

```python
mean_absolute_error(y_test,y_pred3)
```

```python
y_pred3
```

```python
from sklearn.neighbors import KNeighborsRegressor

model5 = KNeighborsRegressor(n_neighbors=2)
model5.fit(x_train, y_train)
y_pred5 = model5.predict(x_test)
```

```python
model5.score(x_train,y_train)
```

```python
mean_absolute_error(y_test,y_pred5)
```
