```python
import pandas as pd
import numpy as np
import os
import glob
from datetime import datetime
import xml.etree.ElementTree as ET 
import cv2
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
annotations_dataset='/kaggle/input/face-mask-detection/annotations'
images_dataset='/kaggle/input/face-mask-detection/images'
```

```python
import os
annot=os.listdir(annotations_dataset)
annotations=[]
for link in annot:
    link2=os.path.join(annotations_dataset,link)
    annotations.append(link2)
```

```python
csv2=[]
def xmltocsv(param):
    tree=ET.parse(param)
    root=tree.getroot()
    filename=root[1].text
    width=root[2][0].text
    height=root[2][1].text
    depth=root[2][2].text
    segmented=root[3].text
    name=root[4][0].text
    pose=root[4][1].text
    truncated=root[4][2].text
    difficult=root[4][3].text
    xmin=root[4][5][0].text
    ymin=root[4][5][1].text
    xmax=root[4][5][2].text
    ymax=root[4][5][3].text
    hello={
    'file_id':filename,
    'width':width,
    'height':height,
    'depth':depth,
    'segmented':segmented,
    'name':name,
    'pose':pose,
    'truncated':truncated,
    'difficult':difficult,
    'xmin':xmin,
    'xmax':xmax,
    'ymin':ymin,
    'ymax':ymax
    }
    csv2.append(hello)
for i in range(853):
    parame=annotations[i]
    xmltocsv(parame)
```

```python
cs=pd.DataFrame(csv2)
cs
```

```python
drop=['depth','segmented','pose','truncated','difficult']
csv=cs.drop(drop,axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
csv['name']=le.fit_transform(csv['name'])
```
